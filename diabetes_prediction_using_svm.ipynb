```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
```

```python
diabetes_data= pd.read_csv('/content/diabetes_prediction_dataset.csv.zip'
                           )
```

```python
diabetes_data.head()
```

```python
diabetes_data['diabetes'].value_counts()
```

```python

diabetes_data.copy()
```

```python
diabetes_data['gender'].unique()
```

```python
#ordinal encoding
enc= OrdinalEncoder()
```

```python
diabetes_data[['gender']]=enc.fit_transform(diabetes_data[['gender']])
```

```python
diabetes_data.head()
```

```python
diabetes_data['smoking_history'].unique()
```

```python
smoking_history=['former','No Info','current','ever','not current','never']
enc=OrdinalEncoder(categories=[smoking_history])
diabetes_data[['smoking_history']]=enc.fit_transform(diabetes_data[['smoking_history']])
```

```python
diabetes_data.head()
```

```python
#seperating data labels
X=diabetes_data.drop(columns='diabetes',axis=1)
Y=diabetes_data['diabetes']
```

```python
print(X)
```

```python
# standardize
scaler=StandardScaler()
scaler.fit(X)
standardized_data=scaler.transform(X)
```

```python
print(X)
print(Y)
```

```python
print(standardized_data)
```

```python
X=standardized_data
```

```python
#training and testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=1,stratify=Y)
```

```python
# traing model
classifier=svm.SVC(kernel='linear')
classifier.fit(X_train,Y_train)
```

```python
# accuracy on training data
X_train_prediction=classifier.predict(X_train)
accuracy_training_data=accuracy_score(X_train_prediction,Y_train)
print(accuracy_training_data)
```

```python
# accuracy on test data
X_test_prediction=classifier.predict(X_test)
accuracy_test_data=accuracy_score(X_test_prediction,Y_test)
print(accuracy_test_data)
```
